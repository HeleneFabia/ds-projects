# Visualizing Loans Awarded by Kiva

Data from <a href = "https://www.kaggle.com/fkosmowski/kivadhsv1" target = "_blank">Kaggle</a>. The dataset contains information about loans awarded by the non-profit <a href = "https://www.kiva.org/" target = "_blank">Kiva</a>. 

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('kiva_data.csv')
df.head()
```

```python
df.head(25)
```

## Bar Charts

```python
import matplotlib.ticker as mtick

f, ax = plt.subplots(figsize=(15, 10))

sns.set_palette('Dark2')
sns.set_context('talk')
sns.barplot(data=df, x="country", y = "loan_amount")

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

plt.title('Kiva Loans Granted per Country')
plt.ylabel("Loans Amount")
plt.xlabel("Country")
```

```python
f, ax = plt.subplots(figsize=(15, 10))

sns.barplot(data=df, x="country", y="loan_amount", hue = 'gender')

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

plt.title('Kiva Loans Granted per Country and Gender')
plt.ylabel("Loans Amount")
plt.xlabel("Country")
```

## Box Plots

```python
f, ax = plt.subplots(figsize=(15, 10))

sns.boxplot(data = df, x = 'country', y = 'loan_amount')

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

plt.title('Kiva Loans Granted per Country')
plt.ylabel("Loans Amount")
plt.xlabel("Country");
```

```python
f, ax = plt.subplots(figsize=(15, 10))

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

sns.boxplot(data = df, x = 'activity', y = 'loan_amount')

plt.title('Kiva Loans Granted per Activity')
plt.ylabel("Loans Amount")
plt.xlabel("Activity");
```

## Violin Plots

```python
f, ax = plt.subplots(figsize=(15, 10))
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

sns.violinplot(data=df, x="activity", y="loan_amount")

plt.title('Kiva Loans Granted per Activity')
plt.ylabel("Loans Amount")
plt.xlabel("Activity");
```

```python
f, ax = plt.subplots(figsize=(15, 10))

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

sns.violinplot(data = df, x = 'country', y = 'loan_amount')

plt.title('Kiva Loans Granted per Country')
plt.ylabel("Loans Amount")
plt.xlabel("Country");
```

```python
f, ax = plt.subplots(figsize=(15, 10))

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

sns.violinplot(data = df, x = 'country', y = 'loan_amount', hue = 'gender', split = True)

plt.title('Kiva Loans Granted per Country and Gender')
plt.ylabel("Loans Amount")
plt.xlabel("Country");
```
